```python
# Check for missing values in 'Y' and handle them (e.g., remove or impute)
print(Y.isnull().sum())  # Check the number of missing values

# Option 1: Remove rows with missing values in 'Y'
new_dataset.dropna(subset=['is_fraud'], inplace=True)

# Option 2: Impute missing values (e.g., with the mean)
# from sklearn.impute import SimpleImputer
# imputer = SimpleImputer(strategy='mean')
# new_dataset['is_fraud'] = imputer.fit_transform(new_dataset[['is_fraud']])

# Recreate X and Y after handling missing values
X = new_dataset.drop(columns='is_fraud', axis=1)
Y = new_dataset['is_fraud']

# Now perform the train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
